```python
def remove (word):
    for i in word:
        if not i.isalpha():
            word = word.strip(i)
    return word

def getnumber(t):
    return t[1]


file = open(r"C:\Users\Noble Desktop\Downloads\article.txt", "r", encoding="utf8")

file

data = file.read().lower().split()
d = {}
for i in data:
    i = remove(i)
    if i in d:
        d[i] = d[i] + 1
    else:
        d[i] = 1
alist = list(d.items())
alist.sort(key=getnumber, reverse=True)
```
